```python
# 결정트리 자동차 오토바이 자전거
from sklearn import tree
x = [[2,0,1], [2,0,0],[4,1,1],[4,0,1]]
y = [1,2,0,0]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x,y)
```

```python
# 바퀴가 두 개고, 크기가 작으면?
clf.predict([[2,1,1]])
```

```python
# 바퀴가 세개고 크기가 작으면?
clf.predict([[3,1,0]])
```

```python
# 결정트리 바이러스
from sklearn import tree

x = [[0,0], [1,1], [1,0]]
y = [0, 1, 0] 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)

clf.predict([[1.,0.,]])    # 새로운 데이터에 대한 예측 
clf.predict([[0.7,0.6,]])   # 새로운 데이터에 대한 예측
```
