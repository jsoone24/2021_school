```python
# 간단한 plot 차트 그리기
import matplotlib.pyplot as plt

values = [10,30,20,40,30]
plt.plot(values)
plt.show()

items = ['A', 'B', 'C', 'D', 'E']
plt.bar(items, values)
plt.show()

# 간단한 pie 차트 그리기
import matplotlib.pyplot as plt

lang = ['C','java','python']
count = [25,30,45]

plt.pie(count, labels = lang)
plt.show()
```

```python
# 차트 표현 기본 방식
# 1. 라이브러리 임포트
import matplotlib.pyplot as plt

# 2. 데이터 생성
data = [1,3,2,4]

# 3. 차트 생성
plt.plot(data)

# 4. 데이터 생성
plt.show()

# 5. 값 변경
기온변화 = [23,30,33,27,30]
plt.plot(기온변화)
plt.show()

# 6. 점(marker) 표시하기
plt.plot(기온변화, marker = 'o')
plt.show()

# 7. 색상 변경하기
plt.plot(기온변화, marker = 'o', color = 'g') # r,g,b
plt.show()

# 8. 격자 표시하기
plt.plot(기온변화, marker = 'o', color = 'g')
plt.grid( True )
plt.show()

# 9, 제목 표시하기
plt.plot(기온변화, marker = 'o', color = 'g')
plt.grid(True)
plt.title('Temperature')
plt.show()

# 10. 파일로 저장하기
plt.plot(기온변화, marker = 'o', color = 'g')
plt.grid(True)
plt.title('Temperature')
plt.savefig('기온변화.png')
plt.show()
```

```python
# 단계별로 기본 옵션 배우기
import matplotlib.pyplot as plt
import numpy as np

부서A판매량 = [10,30,20,40]
부서B판매량 = [20,35,27,32]

plt.title('Sales Volume comparison')

plt.plot(부서A판매량, marker = 'o', color = 'g')
plt.plot(부서B판매량, marker = 'o', color = 'r')
plt.grid(True)

plt.show()

# 단계별로 고급 옵션 배우기
temp1994 = [39.6, 37.9, 37.9, 34.9]
temp2018 = [29.3, 33.9, 32.5, 32.4]

plt.plot(temp1994, marker = 'o', label = '1994') # marker o, v, ^, s, d, D, +, .
plt.plot(temp2018, marker = 'o', label = '2018')

plt.ylim(25,45) # y축 범위 제한
plt.xlim(0,9) # x축 범위 제한
plt.legend() # 범례 표시
plt.rc('font', family = 'Malgun Gothic') # 한글 제목 달기
plt.title("최고 기온 비교")
plt.grid(True) # 격자 표시
plt.xlabel("날짜 변화") # x축 제목
plt.ylabel("최고 기온") # y축 제목
plt.xticks(np.arange(0,10,1), ('1일','2일','3일','4일','5일','6일','7일','8일','9일','10일')) # X축 값의 표시 단위
plt.yticks(np.arange(25, 42, 1)) # Y축 값의 표시 단위
plt.show()
```
