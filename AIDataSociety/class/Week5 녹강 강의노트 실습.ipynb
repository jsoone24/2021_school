# 데이터 불러오기

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# IRIS Data
iris = sns.load_dataset('iris')

# Tips Data
tips = sns.load_dataset('tips')

# Titanic Data
titanic = sns.load_dataset('titanic')
titanic.describe()

# Flight Data
flights = sns.load_dataset('flights')
flights

# fmri Data
fmri = sns.load_dataset('fmri')
fmri.head()

fmri.info()
```

# Rel-Plot 함수 기초 활용

```python
sns.relplot(data = iris, x = 'sepal_length', y = 'petal_length')
sns.set_style('whitegrid')
sns.relplot(data = iris, x = 'sepal_length', y = 'petal_length', hue = 'species', kind = 'line');
sns.set_style('white')
sns.set_style('whitegrid')
sns.set_style('dark')
sns.set_style('darkgrid')
sns.set_style('ticks')


#스스로 해보기
#Tips 데이터에 대하여 total_bill 컬럼과 tip 컬럼에 대한 관계를 차트로 분석하시오.
# • 위 분석에서 흡연여부(smoker 컬럼)에 따라 색상을 변경하시오.
# • 라인 차트로 표현하시오.

# Relplot기본 활용 2
# 두 개의 수치형(numeric) 변수(컬럼)에 대한 관계 분석
sns.relplot(data=tips, x = 'total_bill', y = 'tip');
# 선형 차트, 분산 영역
sns.relplot(data=tips, x = 'total_bill', y = 'tip', kind = 'line');
```

# 3. Line-Plot 함수 (lineplot) 기초 활용

```python
# fmri 데이터에서의 lineplot 적용
sns.lineplot(data=fmri, x = 'timepoint', y ='signal');

# titanic 데이터에서의 lineplot 적용
sns.lineplot(data=titanic, x = 'age', y = 'fare');
sns.relplot(data=titanic, x='age',y = 'fare',kind='line')

```

# Dist-Plot 함수 (distplot) 기초 활용

```python
import pandas as pd
import seaborn as sns

#승객들의 탑승 요금 분석
승객데이터 = pd.read_excel('승객.xlsx')
sns.distplot(승객데이터.fare);
#승객들의 탑승 요금 분석 (필터링)
승객데이터 = pd.read_excel('승객.xlsx')
필터승객=승객데이터.query('fare<=150')
sns.distplot(필터승객.fare);
#승객들의 나이 분석
승객데이터 = pd.read_excel('승객.xlsx')
sns.set_style('whitegrid')
sns.distplot(승객데이터.age.dropna(), bins=20);
#히스토그램 차트 : Distribution
sns.distplot(iris.sepal_length);
sns.distplot(iris.sepal_length, bins=20);
```

# 5. Count-Plot 함수 (countplot) 기초 활용

```python
# 1개의 범주형 변수 분석
sns.countplot(titanic.sex);
sns.countplot(titanic.who);
```

# 6. Bar-Plot 함수 (barplot) 기초 활용

```python
# 성별 생존률 분석
sns.barplot(data=titanic, x = 'who', y = 'survived');
plt.show()
# 좌석 등급별 생존률 분석
sns.barplot(data=titanic, x = 'class', y = 'survived');
plt.show()

# iris data
sns.barplot(data=iris, x = 'species', y = 'sepal_length');
plt.show()
# 수평 막대 차트
sns.barplot(data=iris, y = 'species', x = 'sepal_length', orient='h'); #'v' 수직. xy축 위치 변경해야
plt.show()

# 월별 항공사 승객수
sns.barplot(data=flights, x = 'month', y = 'passengers');
plt.show()
sns.barplot(data=flights, x = 'month', y = 'passengers', orient='v');
plt.show()
plt.figure(figsize=(12,6))
plt.grid()
sns.barplot(data = flights, x = 'month', y='passengers');
plt.show()
```
