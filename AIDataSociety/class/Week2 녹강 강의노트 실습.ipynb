```python
# 강의노트 2주 2강
# 오토바이, 자동차, 자전거 구별 외우지마라
from sklearn import tree
# [바퀴개수, 크기[작다0, 크다1], 동력[사람0, 엔진1]]
x = [[2,0,1], [2,0,0], [4,1,1], [4,0,1]]
# [0, 1, 2] 자동차, 오토바이, 자전거
y = [1, 2, 0, 0] 
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)

from sklearn.tree import export_graphviz
export_graphviz(clf, 
                out_file='tree.dot', 
                class_names=['car', 'auto', 'bike'], 
                impurity=False, 
                filled=True)

import graphviz
with open('tree.dot') as f : 
  dot_graph = f.read()
display(graphviz.Source(dot_graph))

# 새로운 데이터에 대한 예측
clf.predict([[2,1,1]])  

# 새로운 데이터에 대한 예측
clf.predict([[3,1,0]])   




# 바이러스 양성 음성 검사
from sklearn import tree

# 데이터 생성
# [증상1[있으면 1, 없으면 0], 증상2[있으면 1, 없으면 0]]
x = [[0,0], [1,1], [1,0]]
# [바이러스 유무]
y = [0, 1, 0] 
# 모델 생성
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)

# 새로운 데이터에 대한 예측 
clf.predict([[1.,0.,]])    
# 새로운 데이터에 대한 예측 
clf.predict([[0.7,0.6,]]) 

from sklearn.tree import export_graphviz
export_graphviz(clf, 
                out_file='tree.dot', 
                class_names=['음성', '양성'], 
                impurity=False, 
                filled=True)

import graphviz
with open('tree.dot') as f : 
  dot_graph = f.read()
display(graphviz.Source(dot_graph))

# 그림에서 Samples는 데이터게수
# Class는 샘플들이 많이 차지하는 타겟 클래스
# Value는 입력 속성들을 합한 값(리스트)
```
