```python
import csv
import matplotlib.pyplot as plt

f = open('p.csv', encoding='cp949')
d = csv.reader(f)
next(d)
temp = []
temp2 = []
for r in d:
    if '잠실4동' in r[0]:
        for i in range(3,104):
            temp.append(int(r[i].replace(',','')))
    if '상도제1동' in r[0]:
        for i in range(3,104):
            temp2.append(int(r[i].replace(',','')))

plt.plot(temp)
plt.plot(temp2)
plt.show()

'''
for k in d:
    print(k)
'''
```
